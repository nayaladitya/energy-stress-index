# CESI: walkthrough

The Civilisational Energy Stress Index is a ratio of an aggregate demand composite D(t) to a thermodynamic supply capacity S(t), indexed so that 1980 = 100. Its purpose is not prediction but *mechanism separability*: decomposing long-run energy-system constraint into EROI, reserves, and demand so that each can be held fixed and the counterfactual trajectory can be read off.

This notebook loads the backtest output, plots the 1980–2023 series, and shows the one result the paper is built around.

## Formulas

**CESI(t) = [ D(t) / S(t) ] × 100**

**D(t) = ( w_E · Ẽ(t) + w_X · X̃(t) + w_I · Ĩ(t) + w_P · P̃(t) ) × 100**

**S(t) = [ (R/P)_eff(t) / (R/P)_eff(1980) ] · [ EROI_eff(t) / EROI_eff(1980) ] × 100**

with super-linear penalties applied below R/P = 20 years and below EROI = 7.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../results/series/cesi_backtest_data.csv')
df.head()
```

```python
fig, ax = plt.subplots(figsize=(9, 4.5))
ax.plot(df['year'], df['CESI'], lw=2.0, color='#1F3A5F')
ax.set_xlabel('Year')
ax.set_ylabel('CESI (1980 = 100)')
ax.set_title('CESI, 1980–2023')
for y, label in [(1986, '1986 oil glut'), (2008, '2008'), (2014, '2014'), (2020, 'COVID')]:
    ax.axvline(y, color='grey', lw=0.6, ls=':')
    ax.text(y + 0.2, ax.get_ylim()[1] * 0.92, label, fontsize=8, color='grey')
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## The one insight

Civilisational-scale EROI fell from roughly 25:1 in 1980 to 9:1 in 2023. In the causal-isolation projection to 2050 (Section 12 of the paper), freezing EROI at its 2023 value removes approximately 88% of the Central-scenario rise. Freezing reserves has essentially no effect. Freezing demand removes about 56%.

This is why the paper argues that the binding long-run energy constraint is thermodynamic (the falling efficiency of extraction) rather than volumetric (the quantity of resource remaining). It is a different kind of problem to solve.
